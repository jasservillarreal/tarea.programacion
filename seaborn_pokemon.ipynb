```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import pandas as pd
import zipfile

df = pd.read_csv(r"C:\Users\jasser villarreal\Desktop\corte2pro\pokemon.csv")
df.head(10)
```

```python
len(df)
```

```python
##con matplotlib
plt.plot(df.Attack, df.Defense, 'bo')
plt.xlabel('Attack')
plt.ylabel('Type 1')
plt.show() 
#no hay una correlación fuerte entre el ataque y la defensa.
```

```python
#con seaborn
sns.lmplot(x='Attack', y='Defense', data=df)
plt.show()
#gráfico de regresión lineal con seaborn
```

```python
#con seaborn

sns.set_style('darkgrid')
sns.lmplot(x='Attack', y='Defense', data=df, fit_reg = False, hue='Type 1')
plt.title("Ataque y Defensa de un pokemón")
plt.show() 
# gráfico de dispersión con seaborn
```

```python
plt.figure(figsize=(12,9))
sns.boxplot(data=df)
# crea un diagrama de caja (boxplot) de todas las columna
```

```python
plt.figure(figsize=(14,8))
sns.set_style('darkgrid')
sns.violinplot(hue='Type 1', y='Attack', data=df, palette='Set3') 
plt.show()
#sns.set_style('darkgrid') cambiar fondo
# crea grafico de violin  con seaborn
```

```python
sns.barplot(hue='Type 1', y='Attack', data=df,palette='Set2')
#plt.legend(loc='upper right')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
```

```python
sns.distplot(df['Attack'])
#distribucion variable
```

```python
sns.kdeplot(df['Attack'])
#Este gráfico es útil para visualizar la distribución de una variable.
```

```python
sns.jointplot(x='Attack', y='Defense', data=df)
# Este gráfico es útil para visualizar la relación entre dos variables y trazar una línea de regresión.
```

```python
sns.pairplot(df)
#Este gráfico es útil para visualizar las relaciones por pares entre múltiples variables.
```

```python
sns.swarmplot(x='Type 1', y='HP', data=df, hue='Type 2', palette='Set2')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')



#Este gráfico es útil para visualizar la distribución de una variable categórica.
```
