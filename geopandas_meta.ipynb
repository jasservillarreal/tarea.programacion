```python
import geopandas as gpd
import matplotlib.pyplot as plt
```

```python
departamentos = gpd.read_file(r"C:\Users\jasser villarreal\Desktop\corte2pro\Departamentos.zip")
departamentos.head(20)
```

```python
print(departamentos.columns)
```

```python
departamentos.rename(columns={'DeNombre': 'nombre'}, inplace=True)
departamentos
```

```python
departamentos = departamentos.set_index("nombre")
```

```python
departamentos
```

```python
departamentos.plot()
```

```python
departamentos["area"] = departamentos.area
departamentos["area"]
```

```python
departamentos["perimetro"] = departamentos.boundary
```

```python
departamentos["centroide"] = departamentos.centroid



```

```python
departamentos
```

```python
punto_inicial = departamentos["centroide"].iloc[0]
```

```python
punto_inicial
```

```python
departamentos["distancia"] = departamentos["centroide"].distance(punto_inicial)
departamentos["distancia"]
#punto iniciial huila
```

```python
departamentos["distancia"].describe()
```

```python
departamentos.plot(column="area", legend = True,figsize=(10,12))
#, legend = True,figsize=(40,22), legend_kwds = {"loc": "upper left"}
```

```python
departamentos_a= departamentos.set_geometry("centroide")
##mi nueva geometria  para esta variable
#perimetro 
#centroide
```

```python
departamentos_a.plot()
```

```python
departamentos["envolvente"] = departamentos.convex_hull
departamentos["envolvente"]
```

```python
axis = departamentos["envolvente"].plot(alpha = 0.5)
departamentos["perimetro"].plot(ax = axis, color = "black", linewidth = 0.2)
```

```python
rios = gpd.read_file(r"C:\Users\jasser villarreal\Desktop\corte2pro\DrenajeDoble.zip")
rios
```

```python
rios.plot(color="black")
```

```python
colegios = gpd.read_file(r"C:\Users\jasser villarreal\Desktop\corte2pro\EstablecimientosEducativos.zip")
colegios.head(5)
```

```python
Meta = departamentos.query("nombre == 'Meta'")
#condicion
Meta
```

```python
Meta.plot()
```

```python
rios_Meta = gpd.overlay(Meta, rios, how = "intersection")
rios_Meta.plot(figsize = (20, 12))
#geometria
```

```python
axis = Meta.plot(figsize = (20, 12), alpha = 0.2)
rios_Meta.plot(ax = axis, color = "red")
```

```python
colegios_Meta= colegios.query("DeNombre =='META'")
#condicion
```

```python
colegios_Meta.head(5)
```

```python
colegios_Meta.plot()
```

```python
axis = Meta.plot(figsize = (20, 12), alpha = 0.5)
colegios_Meta.plot(ax = axis, color = "white")
rios_Meta.plot(ax = axis, color = "red")
```

```python
rios_Meta.crs
#nombre_nuevo= nombre.to_crs("EPSG:4326")
#codigo del sistema de referencia mas comun EPSG:4326  latitud y longitud
# cartografía a nivel nacional en Colombia=EPSG:9377
```

```python
colegios_Meta["buffer0"] = colegios_Meta.buffer(5000)
#area de influencia
colegios_Meta["buffer0"].plot(figsize = (20, 12))
#generalmente ocurre cuando intentas modificar un DataFrame mientras se hace referencia a una parte de él,
```

```python
rios_Meta["buffer"] = rios_Meta.buffer(500)
 #area de influencia

rios_Meta["buffer"].plot(figsize = (20, 12))
```

```python
rios_Meta = rios_Meta.set_geometry("buffer")
#mi nueva geometria  devolver geometria
```

```python
colegios_afectados = gpd.overlay(colegios_Meta, rios_Meta, how = "intersection")
colegios_afectados.head()
```

```python
axis = colegios_Meta.plot(figsize = (20, 12),color="green")
colegios_afectados.plot(ax = axis, color = "red")
rios_Meta.plot(ax = axis, color = "blue")
```
